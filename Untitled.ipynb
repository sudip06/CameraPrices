```python
df2=df.groupby(['Max resolution']).agg({'Price':['mean'],'Max resolution':['count']})
level0=df2.columns.get_level_values(0)
level1=df2.columns.get_level_values(1)
print (level0)
print (level1)
df2.columns=level0+"_"+level1
df2.reset_index(inplace=True)
```

```python
df2.columns
_,_= plt.subplots(figsize=(7, 10))
#sns.jointplot(x='Max resolution', y='Price_mean', data=df2,kind="kde")
sns.barplot(x='Max resolution', y='Price_mean',data=df2)
plt.show()
```

```python
df2=df.groupby(['Effective pixels','Release date']).agg({'Price':['mean'],'Effective pixels':['count']})
level0=df2.columns.get_level_values(0)
level1=df2.columns.get_level_values(1)
print (level0)
print (level1)
df2.columns=level0+"_"+level1
df2.reset_index(inplace=True)

_,_= plt.subplots(figsize=(7, 5))
#sns.jointplot(x='Max resolution', y='Price_mean', data=df2,kind="kde")
sns.barplot(x='Effective pixels', y='Price_mean',hue='Release date',data=df2)
plt.show()
```
